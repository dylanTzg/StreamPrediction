### Analyse De Corrélation :


```python

# Importing the libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the file
df = pd.read_csv('../data.csv', encoding='latin1')
# print(df.head())
print(df)
```

```python
# Convertir la colonne 'streams' en un type de données numérique
df['streams'] = pd.to_numeric(df['streams'], errors='coerce')

# Vérifier s'il y a des données manquantes dans la colonne 'streams'
print(df['streams'].isnull().sum())

# Calculer la popularité moyenne des chansons par année de sortie
popularity_by_year = df.groupby('released_year')['streams'].mean()

# Création du graphique à barres
plt.figure(figsize=(10, 6))
popularity_by_year.plot(kind='bar', color='skyblue')

# Ajout de titres et de labels
plt.title('Popularité des chansons par année de sortie')
plt.xlabel('Année de sortie')
plt.ylabel('Popularité moyenne (Streams)')
plt.xticks(rotation=45)  # Rotation des étiquettes de l'axe x pour une meilleure lisibilité
plt.grid(axis='y')

# Affichage du graphique
plt.tight_layout()
plt.show()
```
